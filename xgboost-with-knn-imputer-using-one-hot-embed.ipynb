```python

import time
import xgboost as xgb
from sklearn.model_selection import train_test_split
start = time.time()  
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test= train_test_split(train_data.drop(['target'], axis=1), train_data['target'], test_size=0.15, random_state=42)

globals()["xgb_5000_{}".format('CS200007')] = xgb.XGBClassifier(silent=False, 
                              booster='gbtree',
                              scale_pos_weight=4,
                              learning_rate=0.05,  
                              colsample_bytree = 1,
                              subsample = 0.75,
                              eval_metric='logloss',
                              objective='binary:logistic', 
                              n_estimators=150, 
                              max_depth=15, 
                              gamma=1, 
                              seed=777)

eval_set = [(X_test, y_test)]

globals()["xgb_5000_{}".format('CS200007')].fit(X_train,y_train, eval_set=eval_set, verbose=True)

xgb_pred = globals()["xgb_5000_{}".format('CS200007')].predict(train_data.drop(['target'], axis=1))


print("time :", time.time() - start)  # 현재시각 - 시작시간 = 실행 시간
```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(train_data['target'],  xgb_pred)
metrics.auc(fpr, tpr)
```
